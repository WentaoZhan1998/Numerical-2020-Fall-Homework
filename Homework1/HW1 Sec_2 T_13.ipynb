```python
import torch 
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.linalg import lu
n=100+1
A=torch.zeros(n)
B=torch.zeros(n)
A[0]=1
B[0]=0.5
for i in range(1,n):
    A[i]=pow(-1,i)*(1/(2*i+1))
    B[i]=B[i-1]*0.5*0.5*((2*i-1)*(2*i-1)/(2*i)/(2*i+1))
```

```python
sumA=torch.zeros(n)
sumB=torch.zeros(n)
sumA[0]=A[0]
sumB[0]=B[0]
for i in range (1,n):
    sumA[i]=sumA[i-1]+A[i]
    sumB[i]=sumB[i-1]+B[i]

sumA=4*sumA
sumB=6*sumB
```

```python
fig, ax1 = plt.subplots(figsize=(7,5))
ax1.set_ylabel(r"$Error$", fontsize=18)
ax1.set_xlabel(r"$Iteration$", fontsize=18)
ax1.plot(torch.abs(np.pi-sumA), label=r"First")
ax1.plot(torch.abs(np.pi-sumB), label=r"Second")
ax1.legend(fontsize=12, loc=0)
ax1.set_yscale("log")
```
